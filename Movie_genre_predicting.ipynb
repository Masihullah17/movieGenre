**Movie Genre**

Using my trained model and weights

```
from keras.models import model_from_yaml
# load YAML and create model
yaml_file = open('movie_genre_final.yaml', 'r')
loaded_model_yaml = yaml_file.read()
yaml_file.close()
model = model_from_yaml(loaded_model_yaml)
# load weights into new model
model.load_weights("movie_genre_final.h5")
print("Loaded model from disk")
```

```
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
train = pd.read_csv('train.csv')    # reading the csv file

img = image.load_img('avengers.jpeg',target_size=(400,400,3))
img = image.img_to_array(img)
img = img/255

classes = np.array(train.columns[2:])
proba = model.predict(img.reshape(1,400,400,3))
top_3 = np.argsort(proba[0])[:-4:-1]
for i in range(3):
    print("{}".format(classes[top_3[i]])+" ({:.3})".format(proba[0][top_3[i]]))
plt.imshow(img)
```
